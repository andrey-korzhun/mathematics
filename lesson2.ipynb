### Задание 1
#### Напишите уравнение параболы, проходящей через три точки (x,y):<br>(1,2), (3,10), (5,1)

Уравнение параболы имеет общий вид: $$ ax^2+bx+c=y $$
Нам извстны x и y в трех точках. Можем составить систему уравнений и решая ее найти коэффициенты a, b и c:

 \begin{cases}
  a+b+c=2, 
   \\
  9a+3b+c=10,
  \\
  25a+5b+c=1
 \end{cases}

#### Решение:


##### Ответ: Искомое уравнение имеет вид:<br>  $$ y = -2,125x^2+12,5x-8,375 $$
#### Проверка:

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return -2.125*x**2 + 12.5*x - 8.375

xlist = np.linspace(0.5, 5.5, 100)
ylist = [func (x) for x in xlist]

x = [1, 3, 5]
y = [2, 10, 1]

plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.plot(xlist, ylist)
plt.grid()
```
