### Задание 2.2
Решите систему уравнений:

Линейная или нелинейная это система?<br>
А каждое уравнение по отдельности?

#### Решение
Данная система уравнений является не линейной.<br>
Первое уравнение системы - квадратное.<br>
Выразим из второго уравнения y через x и подставим в первое


Теперь, подставив x в любое уравнение, найдём y


#### Проверка

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 + y*x - 9, x -y/5)

fsolve(equations, (1, 1))
```

### Задание 3

Решите задачу:<br>
Площадь пола прямоугольной комнаты равна 48 м2,а его периметр равен 28 м. Найдите длину и ширину комнаты.

#### Решение

Составляем систему уравнений:<br>


Выразим из первого уравнения a через b и подставим во второе уравнение:


Ответ: Длина - 8м, ширина - 6 м.

#### Проверка

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x*y - 48, 2*(x + y) - 28)

fsolve(equations, (1, 1))
```

### Задание 4

Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

#### Решение

```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')

x = np.linspace(-2 * np.pi, 2 * np.pi, 1024)

k = 1
plt.figure(figsize=(8, 6))
plt.plot(x, np.cos(k*x))
k -= 0.5
plt.plot(x, np.cos(k*x))
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```
